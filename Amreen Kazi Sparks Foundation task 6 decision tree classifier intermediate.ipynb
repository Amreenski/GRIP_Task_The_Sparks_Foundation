# Import the necessary libraries

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

# Load the iris dataset

```python
iris = load_iris()
```

```python
#Creating a pandas DataFrame from the iris dataset
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
#display the first few rows of the dataset
df.head()
```

```python
#get a summary of the dataset
df.describe()
```

```python
#visualizing the distribution of each feature using histograms
df.hist()
plt.show()
```

```python
#creating a scatter plot matrix to explore the relationships between different features
pd.plotting.scatter_matrix(df, c=df['target'], figsize=(10, 10), marker='o', hist_kwds={'bins': 20}, alpha=0.8)
plt.show()
```

```python
#Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[iris.feature_names], df['target'], test_size=0.3, random_state=42)
```

```python
#Creating an instance of the DecisionTreeClassifier and fit it on the training data
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
#Visualizing the decision tree using the graphviz library
plt.figure(figsize=(12,8))
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```
